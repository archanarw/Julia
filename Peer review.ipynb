### $$ f(x) = 2sin(x)cos(x) $$

```julia
f(x) = 2*sin(x)*cos(x)
g(x) = sin(2*x)
```

### This function returns sin(2x) value, which equals 2*sin(x)*cos(x). I will prove the same by comparing the graphs of the two functions.

```julia
println("x is taken from -2π to 2π,in the interval of π/2.")
println("The values of f(x) for each x is printed below as an ordered pair :")
println("(x,f(x))")
println(" ")
for x = -2*pi:pi/2:2*pi
    println("($x,$(f(x)))")
end
```

```julia
using Plots
gr()
p1 = plot(f,-2*pi,2*pi, title = "This is the 2sin(x)cos(x) graph", lw = 2);
p2 = plot(g,-2*pi,2*pi, title = "This is the sin(2x) graph", lw = 2);
plot(p1,p2,layout = (1,2),legend = false)
```

#### Thus, we see that the two graphs are identical, thereby the two function are the same.
